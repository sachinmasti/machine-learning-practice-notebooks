```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
data = {
    'cgpa': np.random.randint(3,9,300).astype('float'),
    'packge': np.random.randint(2,6,300).astype('float')
}
```

```python
df = pd.DataFrame(data=data)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.compose import TransformedTargetRegressor
from sklearn.preprocessing import FunctionTransformer
```

```python
x = df[['cgpa']]
y = df[['packge']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model = LinearRegression(fit_intercept=True)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(f'mse is :{mean_squared_error(y_test,y_pred)}')
```

```python
print(f'mae is : {mean_absolute_error(y_test,y_pred)}')
```

```python
print(f'r2_score is : {r2_score(y_test,y_pred)}')
```

```python
print(f'rmse is : {np.sqrt(mean_squared_error(y_test,y_pred))}')
```

```python
transformers = FunctionTransformer(np.log1p,np.expm1,validate=True)
regrosor = LinearRegression()
```

```python
model_1 = TransformedTargetRegressor(regressor=regrosor,transformer=transformers)
model_1.fit(x_train,y_train)
```

```python
predictions = model_1.predict(x_test)
```

```python
print(f'mse is : {mean_squared_error(y_test,predictions)}')
```

```python
print(f'mae is : {mean_absolute_error(y_test,predictions)}')
```

```python
print(f'r2_score is : {r2_score(y_test,predictions)}')
```
