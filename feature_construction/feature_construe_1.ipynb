```python
df.drop(columns='show_id',inplace=True)
```

```python
df['type'].value_counts()
```

```python
df['show_type'] = (df['type']=='Movie').astype(int)
```

```python
df.head()
```

```python
df['title_count'] = df['title'].apply(len)
```

```python
df['world_count_by_title'] = df['title'].apply(lambda y: len(y.split()))
```

```python
df.sample(2)
```

```python
df['country_count'] = df.groupby('country')['country'].transform('count')
```

```python
df.info()
```

```python
# df['date_added'] = pd.to_datetime(df['date_added'],format='mixed')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format="%B %d, %Y")
```

```python
df.head(3
)
```

```python
df['day'] = df['date_added'].dt.day
df['month'] = df['date_added'].dt.month_name()
df['year'] = df['date_added'].dt.year
df['day_of_week'] = df['date_added'].dt.dayofweek
```

```python
df['is_weekend'] = df['day_of_week'].isin([5,6]).astype(int)
```

```python
df['release_year']
```

```python
from datetime import  datetime
```

```python
df['content_age'] = datetime.now().year - df['release_year']
```
