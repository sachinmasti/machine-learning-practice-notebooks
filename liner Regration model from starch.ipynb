```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class mera_model():

    def __init__(self):
        self. m = 0
        self.b = 0

    def fit(self,x_train,y_train):
        numenitor = 0
        denomitor = 0

        x_train = x_train.values.ravel()
        y_train = y_train.values.ravel()

        for i in range(x_train.shape[0]):
            numenitor = numenitor + ((x_train[i] - x_train.mean()) * (y_train[i] - y_train.mean()))
            denomitor = denomitor + ((x_train[i] - x_train.mean()) * (x_train[i] - x_train.mean()))

        self.m = numenitor / denomitor
        self.b = y_train.mean() - (self.m * x_train.mean())
        print(self.m)
        print(self.b)

    def predict(self,x_test):
        return self.m * x_test + self.b
```

```python
np.random.seed(42)
```

```python
data = {
    'cgpa': np.random.randint(2,9,300).astype('float'),
    'packge':np.random.randint(2,6,300).astype('float')
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
x= df['cgpa']
y = df['packge']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
x_train
```

```python
model = mera_model()
```

```python
model.fit(x_train,y_train)
model.predict(x_train[[8]])
```

```python
sns.scatterplot(df,x='cgpa',y='packge')
plt.plot(df['cgpa'],model.predict(df[['cgpa']]))
plt.title('best fit line')
```
