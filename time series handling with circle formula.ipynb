```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
import datetime

start_date = datetime.datetime(2025, 1, 1)
end_date = datetime.datetime(2025, 12, 31)
```

```python
data = {
    'age':np.random.randint(18,40,12),
    'month_of_joining': ['jan','feb','march','april','may','june','july','august','september','october','november','december'],
    'salary':np.random.randint(30000,100000,12),
    'salary_credited_time':start_date + datetime.timedelta(days=np.random.randint(0, (end_date-start_date).days), hours=np.random.randint(0,24), minutes=np.random.randint(0,60))
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df['year'] = df['salary_credited_time'].dt.year
```

```python
df['month'] = df['salary_credited_time'].dt.month
df['day'] = df['salary_credited_time'].dt.day
```

```python
joinig_month = {'jan':1,'feb':2,'march':3,'april':4,'may':5,'june':6,'july':7,'august':8,'september':9,'october':10,'november':11,'december':12}
```

```python
df['joing_month'] = df['month_of_joining'].map(joinig_month)
```

```python
df
```

```python
df['months_sin'] = np.sin(2 * np.pi * df['joing_month']/12)
df['month_cos'] = np.cos(2* np.pi * df['joing_month']/12)
```

```python
df['salary_credited_houre'] = df['salary_credited_time'].dt.hour
```
