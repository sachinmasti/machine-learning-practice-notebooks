```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
fig,ax = plt.subplots(figsize=(10,6))
fig.patch.set_facecolor('#b9ebc6')
ax.set_facecolor('#d2d7d9')

ax.plot(df['fare'],color='steelblue',label='fare')

max_val = df['fare'].max()
max_idx = df['fare'].idxmax()

ax.annotate(
    text=f'max : {max_val:.1f}',
    xy=(max_idx,max_val),
    xytext=(max_idx + 20, max_val - 50),
    arrowprops=dict(arrowstyle ='->',color = 'green'),
    fontsize=(11),
    color='#b697db'
)
ax.grid(True,linestyle='--',color='#a9b2cf')
ax.legend(frameon=False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
import mplcyberpunk

plt.style.use("cyberpunk")
fig,ax = plt.subplots(figsize=(8,6))

ax.plot(df['age'],color='steelblue',label='age')
# fig.patch.set_facecolor('#d99cd8')
# ax.set_facecolor('#e3d3e3')

ax.tick_params(axis='both', colors='white')
# ax.tick_params(axis='x',colors='#c9cc8b')

max_val = df['age'].max()
max_idx = df['age'].idxmax()

ax.annotate(
    text = f'max_agg : {max_val:.1f}',
    xy=(max_idx,max_val),
    xytext= (max_idx + 20, max_val - 20),
    arrowprops=dict(arrowstyle = '->', color = 'b'),
    fontsize=10,
    color='red'
)

ax.legend(frameon=False,loc='upper right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title('maximum age',color='red')
mplcyberpunk.add_glow_effects()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import mplcyberpunk

plt.style.use("cyberpunk")

plt.plot([1, 3, 9, 5, 2, 1, 1], marker='o',label='A')
plt.plot([4, 5, 5, 7, 9, 8, 6], marker='o',label='B')

plt.legend()
mplcyberpunk.add_glow_effects()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

scatter = ax.scatter(df['age'], df['fare'], 
                     c=df['fare'],      # ← color kaunse column se aaye
                     cmap='viridis',    # ← color scheme
                     alpha=0.6)

fig.colorbar(scatter, ax=ax, label='Fare Amount')  # ← colorbar lagao

ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_title('Age vs Fare')
plt.show()
```

```python
plt.style.use('cyberpunk')
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0][0].plot(df['age'], color='steelblue')
axes[0][0].set_title('Age')

axes[0][1].hist(df['fare'], color='coral')
axes[0][1].set_title('Fare Distribution')

axes[1][0].scatter(df['age'], df['fare'], alpha=0.5)
axes[1][0].set_title('Age vs Fare')

bars = axes[1][1].bar(range(10), df['fare'].head(10), color='pink')
axes[1][1].set_title('Top 10 Fares')

mplcyberpunk.add_bar_gradient(bars=bars)
mplcyberpunk.add_glow_effects()

plt.tight_layout()
plt.show()
```
