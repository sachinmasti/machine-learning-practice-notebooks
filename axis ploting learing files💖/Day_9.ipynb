```python
import mplcyberpunk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mplcyberpunk as cp
from warnings import filterwarnings

from matplotlib.lines import lineStyles
from plotly.graph_objs.waterfall.decreasing import marker

from Day_5 import max_idx

filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
```

```python
plt.style.use('cyberpunk')
fig, ax = plt.subplots(figsize=(6,6))

ax.tick_params(axis='x',colors='red')
ax.plot(df['fare'],color='pink')
ax.set_xlim(0,100)
ax.set_ylim(1,100)
ax.set_title('fare plot')
```

```python
fig,ax = plt.subplots(figsize=(6,6))

ax.plot(df['fare'],color='yellow')

ax.axhline(y=100,linestyle='--',color='pink',label='y lie')
ax.legend()
```

```python
fig,ax = plt.subplots(figsize=(6,6))

ax.scatter(df['age'],df['fare'],color='#65bf7d')
ax.axvline(x=50,linestyle='--',color='#4777d1',label='x line')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(df['fare'], color='pink', label='fare')

ax.fill_between(
    x=range(len(df['fare'])),  # x axis
    y1=df['fare'],              # upar ki line
    y2=0,                       # neeche tak bharo
    alpha=0.3,                  # transparency
    color='red'
)

ax.set_title('Fare with Fill')
plt.savefig('sachin_88.svg')
plt.show()
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(10,6))

gs = gridspec.GridSpec(2,2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,:])

ax1.plot(df['age'],color='red')
ax1.set_title('plot no 1')
ax1.set_xlim(1,100)
ax1.set_ylim(1,100)
ax1.axhline(y=40,color='brown',linestyle='--',label='ax line')
ax1.legend()

ax1.fill_between(x=x,
                 y2=0,
                 y1=y,
                 interpolate=True)

sns.barplot(df['survived'].value_counts(),ax=ax2,palette=['pink','steelblue'])
ax2.set_title('survived wise bar plot')

max_idx = df['fare'].idxmax()
max_val = df['fare'].max()
ax3.plot(df['fare'],color='yellow',marker='o')

ax3.annotate(text=f'{max_val:.2f}',
             xy=(max_idx,max_val),
             xytext=(max_idx + 30, max_val - 50),
             arrowprops=dict(arrowstyle='->',color='red'),
             fontsize=10,
             color='pink')

ax3.set_title('fare plot')
```
