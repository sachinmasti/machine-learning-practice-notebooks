```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(2, 2)  # 2 row, 2 column ka grid

ax1 = fig.add_subplot(gs[0, :])   # pehli row — poori width!
ax2 = fig.add_subplot(gs[1, 0])   # doosri row — left
ax3 = fig.add_subplot(gs[1, 1])   # doosri row — right

ax1.plot(df['fare'], color='steelblue')
ax1.set_title('Fare — Full Width')

ax2.hist(df['age'].dropna(), color='coral')
ax2.set_title('Age Distribution')

ax3.boxplot(df['fare'].dropna())
ax3.set_title('Fare Outliers')

plt.tight_layout()
plt.show()
```

```python
import mplcyberpunk
import mplcyberpunk
plt.style.use('cyberpunk')
fig = plt.figure(figsize=(10,8))

gs = gridspec.GridSpec(2,2)

ax1=fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,:])

ax1.plot(df['age'],color='pink')
ax1.set_title('age plot')

ax2.hist(df['fare'].dropna(),color='yellow')
ax2.set_title('fare distribution')

ax3.pie(df['alive'].value_counts(),labels=df['alive'].value_counts().index,
                autopct='%1.1f%%',colors=['black','steelblue'],shadow=True)

plt.tight_layout()
mplcyberpunk.add_glow_effects(gradient_fill=True)
```

```python
fig = plt.figure(figsize=(10,8))

gs = gridspec.GridSpec(2,2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[0:,1])

ax1.plot(df['fare'],color='pink',marker='o')
ax1.set_title('fare plot')

scatter = ax2.scatter(df['age'],df['fare'],c=df['fare'],cmap='viridis',alpha=0.5)
fig.colorbar(scatter,ax=ax2,label='fare wise color')


bar = sns.barplot(df['survived'].value_counts(),ax=ax3,palette=['green','yellow'])
mplcyberpunk.add_bar_gradient(bars=bar.containers[0])
```
