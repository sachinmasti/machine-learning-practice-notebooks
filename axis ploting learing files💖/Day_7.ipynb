```python
import mplcyberpunk
import mplcyberpunk
plt.style.use('cyberpunk')

fig,ax = plt.subplots(2,2,figsize=(15,10))

fig.suptitle('all plots at ones',color='#4287f5')

ax[0][0].hist(df['age'].dropna(),color='#7595c9')
ax[0][0].set_title('age distribution',color='green',fontweight='bold')
ax[0][0].set_xlabel('age',color='#83ebad')
ax[0][0].set_ylabel('count',color='#83ebad')

ax[0][1].boxplot(df['fare'].dropna())
ax[0][1].set_title('fare outliers',color='#bbdb5a')
max_val=df['fare'].max()
ax[0][1].annotate(
            text=f'{max_val:.2f}',
            xy=(1,max_val),
            xytext=(1.25,max_val - 50),
            arrowprops=dict(arrowstyle ='->',color='pink'),
            fontsize=8,
            color='red'
)

scatter = ax[1][0].scatter(df['age'],df['fare'],c=df['fare'],alpha=0.5,cmap='viridis')

fig.colorbar(scatter,ax=ax[1][0],label='fare by age')

bar = df['survived'].value_counts().plot(kind='bar',ax=ax[1][1],color=['#a11b8d','#6f3cd6'])
ax[1][1].set_title('survived')
ax[1][1].tick_params(rotation=0)

mplcyberpunk.add_bar_gradient(bars=bar.containers[0])
mplcyberpunk.add_glow_effects(gradient_fill=True)
```
