```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
from matplotlib import figure
import mplcyberpunk
```

```
df = sns.load_dataset('titanic')
```

```
df.head()
```

```
fig,ax = plt.subplots(figsize=(5,6))

ax.boxplot(df['age'].dropna())
ax.set_xlabel('age outliers')
ax.set_title('boxplot',color='red',fontweight='bold')
```

```
fig,ax = plt.subplots(figsize=(8,6))

ax.boxplot([df['age'].dropna(),df['fare'] ], label=['age', 'fare'])
ax.set_title('age and fare outliers')
```

```
plt.style.use('cyberpunk')
fig,ax = plt.subplots(figsize=(10,6))

ax.violinplot(df['age'].dropna())
ax.set_title('age plot')
mplcyberpunk.add_glow_effects(gradient_fill=True)
```

```
fig,ax = plt.subplots(1,2,figsize=(10,6))

ax[0].boxplot(df['fare'].dropna())
ax[0].set_title('fare box plot')

ax[1].violinplot(df['fare'].dropna())
ax[1].set_title('fare voilen plot')
```
