```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

```python
sns.set_style('darkgrid')
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
fig,ax = plt.subplots()

ax.plot(df['age'],label='age')
ax.plot(df['fare'],label='fare')

ax.legend(loc='best',fontsize=10,frameon=False,title='age vs fare',title_fontsize=15)
```

```python
df.head()
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,6))
fig.patch.set_facecolor('#a4bbe0')
ax[0].set_facecolor('#a0a5ad')
ax[1].set_facecolor('#ead9fa')

ax[0].scatter(df['age'],df['fare'],label='age and fare dist',color='#2a2a3e')

ax[1].plot(df['fare'],label='fare',color='steelblue',linestyle='--',markersize=4,marker='o')

ax[0].legend(loc='upper left',title_fontsize=15,fontsize=10,title='age dist')
ax[1].legend(loc='upper left',frameon=False)

ax[0].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)

fig.tight_layout()
```

```python
plt.style.use('ggplot')
fig,ax = plt.subplots(1,2,figsize=(10,8))
fig.patch.set_facecolor('#8C8B85')
ax[1].set_facecolor('#E3D294')

ax[0].pie(df['class'].value_counts(),labels=df['class'].value_counts().index,autopct='%1.1f%%',shadow=True,explode=[0.1,0,0])
ax[0].set_title('pie chart according to class',fontsize=15,fontweight='bold',color='#B7E8D9')

sns.kdeplot(df['age'],ax=ax[1])
ax[1].set_title('distribution of age',fontsize=15,color='#292227',fontweight='bold')
ax[1].set_xlabel('age',color='#142424',fontsize=15)

ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)

plt.tight_layout()
```
