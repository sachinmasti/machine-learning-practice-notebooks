```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
fig , ax = plt.subplots()

ax.plot(df['age'],linestyle='--',color='steelblue',linewidth=2,marker='o',markersize=5)

ax.grid(True,linestyle='--',alpha=1)

ax.set_title('mere plot',fontsize=9,fontweight='bold',color='green')
ax.set_xlabel('size',fontsize=15,fontweight='bold',color='red')
ax.set_ylabel('age',fontsize=14,fontweight='bold',color='red')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
fig,ax = plt.subplots(figsize=(10,8))

ax.plot(df['fare'],linestyle='--', color='#47c3c9', marker='o' ,markersize=5,linewidth=2)

ax.set_title('fare plot', fontsize=10,fontweight='bold',color='#d63cd4')
ax.set_xlabel('fare amount',fontsize=10,fontweight='bold',color='#d63cd4')
ax.set_ylabel('y axis',fontsize=10, fontweight='bold',color='#d63cd4')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.grid(True,linestyle='--',alpha=0.5,color='#6343b5')
```
