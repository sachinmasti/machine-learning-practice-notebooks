```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
fig, ax = plt.subplots(1, 2)

ax[0].bar(df['age'].head(10),height=range(10),color=['green' if i > 10 else 'yellow' for i in df['age']])
ax[1].hist(df['fare']) 

ax[0].set_title('1st plot',color='red')
ax[1].set_title('2nd plot',color='green')

ax[0].axhline(y=10,color='black',linestyle='--')
```
