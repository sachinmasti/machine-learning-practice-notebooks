```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)  # 2 plots upar neeche

ax[0].plot(df['age'])
ax[1].plot(df['fare'])
# Ab dono ka X axis same — time series mein bahut kaam aata hai!
```

```python
fig,ax = plt.subplots(2,1, sharex=True)

ax[0].plot(df['fare'])
ax[1].plot(df['pclass'])
```

```python
fig,ax = plt.subplots(2,1, sharey=True)

ax[0].plot(df['fare'])
ax[1].plot(df['age'])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
# ↑ aaj ka dono seekha hua ek saath!

ax[0].bar(df['age'].head(10), df['age'].head(10), color='steelblue')
ax[1].hist(df['fare'])

ax[0].set_title('Age')
ax[1].set_title('Fare')

plt.tight_layout()  # bonus! spacing automatically fix ho jaati hai
plt.show()
```
