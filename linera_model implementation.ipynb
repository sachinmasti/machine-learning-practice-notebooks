```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
np.random.seed(42)
```

```python
data = {
    'cgpa':np.random.randint(3,9,300),
    'packge':np.random.randint(2.10,6.50,300).astype('float32')
}
```

```python
df = pd.DataFrame(data=data)
```

```python
import plotly.express as px
```

```python
sns.scatterplot(df,x='cgpa',y='packge')
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = df[['cgpa']]
```

```python
y = df[['packge']]
```

```python
y
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.15,shuffle=True)
```

```python
x_train
```

```python
y_train
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
sns.scatterplot(df,x='cgpa',y='packge')
plt.plot(df['cgpa'],model.predict(df[['cgpa']]),color='green')
plt.title('best fit line from model')
```

```python
m = model.coef_
```

```python
x = model.intercept_
```
