```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pylab as plt
import plotly.express as px
```

```python
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.shape
```

```python
x=df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import SimpleImputer,IterativeImputer
from sklearn.ensemble import VotingClassifier
from sklearn.decomposition import PCA
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.85,shuffle=True,random_state=42)
```

```python
x_train.shape
```

```python
x_train
```

```python
y_train
```

```python
leble = LabelEncoder()
```

```python
y_train_encoded = leble.fit_transform(y_train)
y_test_encoded = leble.transform(y_test)
```

```python
scale = MinMaxScaler()
```

```python
x_train_transformed = scale.fit_transform(x_train)
```

```python
pca = PCA(n_components=None)
```

```python
pca_train = pca.fit_transform(x_train_transformed)
pca_test = pca.transform(x_test)
```

```python
col = [f"pca{i}" for i in range(1,pca_train.shape[1]+1)]
```

```python
pcd = pd.DataFrame(pca_train,columns=col)
pcd['species'] = y_train
```

```python
pcd
```

```python
fig = px.scatter_3d(pcd,
    x='pca1',y='pca2',z='pca3',
    color='species',
    title='pca_3D_plot',
    # color_continuous_scale='RdBu'
)
fig.show()
```

```python
fig = px.line(
    x = range(1,len(pca.explained_variance_ratio_)+1),
    y = pca.explained_variance_ratio_.cumsum(),
    markers = True,
    labels = {'x': 'number of componets','y':'convergence'}
)
fig.add_hline(line_dash='dash',y=0.90,line_color='yellow',annotation_text='90% threshold')
fig.show()
```

```python
model = DecisionTreeClassifier()
```

```python
trined_model = model.fit(pca_train,y_train_encoded)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_pred = model.predict(pca_test)
```

```python
print(f'acurracy is {accuracy_score(y_test_encoded,y_pred)}')
```

```python
confusion_matrix(y_test_encoded,y_pred)
```
